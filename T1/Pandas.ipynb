### Concat

```python
df1=pd.DataFrame({
    "A":[1,2],
    "B":[3,4]
})
df1
```

```python
df2=pd.DataFrame({
    "A":[5,6],
    "B":[7,8]
})
df2
```

```python
pd.concat((df1,df2),axis=0)
```

```python
pd.concat([df1,df2],axis=0,ignore_index=True)
```

```python
pd.concat([df1,df2],axis=0,keys=["df1","df2"])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df3=pd.DataFrame({
    "C":[5,6],
    "B":[7,8]
})
df3
```

```python
pd.concat([df1,df3],axis=0)
```

```python
pd.concat((df1,df3),axis=1)
```

```python
pd.concat((df1,df3),axis=1,sort=True)
```

```python
df4=pd.DataFrame({
    "A":[1,2,3],
    "B":[3,4,5]
})
df4
```

```python
pd.concat([df4,df3],axis=1)
```

```python
pd.concat((df4,df3))
```

```python
pd.concat((df4,df3),join="inner")
```

```python
pd.concat((df4,df3),join="outer")
```

### Merge

```python
df1=pd.DataFrame({
    "name":["VHA","PSP","VBY","DPS","MSS"],
    "sub":["python","fsd","coa","toc","dm"],
    "id":[3,2,1,6,4]
})
df1
```

```python
df2=pd.DataFrame({
    "id":[1,2,3,4,5],
    "marks":[10,9,10,9,9]
})
df2
```

```python
pd.merge(df1,df2,on="id")
```

```python
pd.merge(df1,df2,on="id",how="inner")
```

```python
pd.merge(df1,df2,on="id",how="outer")
```

```python
pd.merge(df1,df2,on="id",how="left")
```

```python
pd.merge(df1,df2,on="id",how="right")
```

```python
pd.merge(df1,df2,how="cross")
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```

```python
df1=pd.DataFrame({
    "name":["VHA","PSP","VBY","DPS","MSS"],
    "sub":["python","fsd","coa","toc","dm"],
    "in":[3,2,1,6,4]
})
df1
```

```python
df2=pd.DataFrame({
    "id":[1,2,3,4,5],
    "marks":[10,9,10,9,9]
    ,"fac":["VHA","NAS","VBY","DPS","MSS"]
})
df2
```

```python
df1.merge(df2,left_on="in",right_on="id")
```

```python
df1.merge(df2,left_on="name",right_on="fac",how="outer")
```
