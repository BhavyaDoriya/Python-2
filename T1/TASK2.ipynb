```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../Datasets/IPL_Matches_2008_2022.csv")
df
```

```python
df.groupby("Season")["WinningTeam"].apply(lambda x:x.value_counts().head(1))
```

```python
df.groupby("WinningTeam")["Player_of_Match"].nunique().sort_values(ascending=False)[0]
```

```python
df1=df.groupby("Venue")["Venue"].value_counts().sort_values(ascending=False).head()
df1=df1.reset_index()
df1
```

```python
df2=df.merge(df1,on="Venue",how="inner")
df2
```

```python
df2.groupby("Venue")["WinningTeam"].apply(lambda x:x.value_counts().head(1))
```
