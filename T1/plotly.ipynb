```python
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../Datasets/social_media_engagement.csv")
df
```

```python
df.nunique()
```

```python
fig=px.bar(df,x="post_type",y="likes",width=500,height=500,title="Likes by Post Type")
# or
# fig=px.bar(x=df.post_type,y=df.likes)
fig.update_layout(xaxis_title="Post Type",yaxis_title="Likes")
fig.show()
```

```python
fig=px.bar(df,x="post_type",y="likes",width=500,height=500,title="Likes by Post Type",text_auto=True)
fig.update_layout(xaxis_title="Post Type",yaxis_title="Likes")
fig.show()
```

```python
fig=px.bar(df,x="post_type",y="likes",width=500,height=500,title="Likes by Post Type",color="platform")
fig.update_layout(xaxis_title="Post Type",yaxis_title="Likes")
fig.show()
```

```python
fig=px.bar(df,x="post_type",y="likes",width=500,height=500,title="Likes by Post Type",color="platform",color_discrete_map={"Facebook":"blue","Instagram":"magenta","Twitter":"cyan"})
fig.update_layout(xaxis_title="Post Type",yaxis_title="Likes")
fig.show()
```

```python
fig=px.bar(df,x="post_type",y="likes",width=900,height=800,title="Likes by Post Type",color="platform",color_discrete_sequence=["blue","magenta","cyan"])
fig.update_layout(xaxis_title="Post Type",yaxis_title="Likes")
fig.show()
```

```python
fig=px.bar(df,x="post_type",y="likes",width=800,height=600,title="Likes by Post Type",color="comments")
fig.update_layout(xaxis_title="Post Type",yaxis_title="Likes")
fig.show()
```

```python
fig=px.bar(df,x="post_type",y="likes",width=800,height=600,title="Likes by Post Type",color="comments",hover_data=["post_day","sentiment_score"])
fig.update_layout(xaxis_title="Post Type",yaxis_title="Likes")
fig.show()
```

```python
fig=px.bar(df,y="post_type",x="likes",width=800,height=600,title="Likes by Post Type",color="comments",hover_data=["post_day","sentiment_score"],orientation="h")
fig.update_layout(xaxis_title="Post Type",yaxis_title="Likes",)
fig.show()
```

```python
fig=px.bar(df,y="post_type",x="likes",width=800,height=600,title="Likes by Post Type",color="comments",hover_data=["post_day","sentiment_score"],orientation="h",text="likes")
fig.update_layout(xaxis_title="Post Type",yaxis_title="Likes",)
fig.show()
```

```python
fig=px.bar(df,x="post_type",y="likes",width=800,height=900,title="Likes by Post Type",color="platform",
           hover_data=["post_day","sentiment_score"],pattern_shape="sentiment_score")
fig.update_layout(xaxis_title="Post Type",yaxis_title="Likes")
fig.show()
```

```python
fig=px.bar(df,x="post_type",y=["likes","shares","comments"],width=800,height=900,title="Engagement by Post Type")
fig.update_layout(xaxis_title="Post Type",yaxis_title="Likes")
fig.show()
```

```python
fig=px.bar(df,x="post_type",y=["likes","shares","comments"],width=800,height=900,title="Engagement by Post Type",barmode="stack")
fig.update_layout(xaxis_title="Post Type",yaxis_title="Likes")
fig.show()
```

```python
fig=px.bar(df,x="post_type",y=["likes","shares","comments"],width=800,height=900,title="Engagement by Post Type",barmode="group")
fig.update_layout(xaxis_title="Post Type",yaxis_title="Likes")
fig.show()
```

```python
df1=df.groupby("platform",as_index=False)[["likes","comments","shares"]].sum()
df1
```

```python
fig=px.pie(df1,names="platform",values="likes",color_discrete_sequence=["pink","beige","lightblue"])
fig.update_traces(textinfo="percent+label",textposition="outside",pull=[0,0,0.2])
fig.show()
```
