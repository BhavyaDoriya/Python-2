### Webscraping with a website

```python
url="https://subslikescript.com/movie/Mamma_Mia-795421"
response=requests.get(url)
response.status_code
```

```python
soup=BeautifulSoup(response.text,"html.parser")
soup
```

```python
article=soup.find("article",class_="main-article")
article
```

```python
# Finding top title on the webpage
title=article.find("h1").text
title
```

```python
# finding script
soup.find("div",class_="full-script")
script=soup.find_all("p")
data=""
for i in script:
    data += i.text+"\n"
data
```

```python
with open(f"txtFiles\{title}.txt","w", encoding="utf-8") as f:
    f.write(data)
```

```python
url="https://subslikescript.com/movies_letter-V"
response=requests.get(url)
response.status_code
```

```python
soup=BeautifulSoup(response.text,"html.parser")
soup
```

```python
parent=soup.find("article",class_="main-article")
parent
```

```python
anchor_tags=parent.find_all("a")
len(anchor_tags)
```

```python
anchor_tags
```

```python
links=[]
for i in anchor_tags:
    links.append("https://subslikescript.com/"+i['href'])
links
```

```python
for link in links:
    response=requests.get(link)
    soup=BeautifulSoup(response.text,"html.parser")
    article=soup.find("article",class_="main-article")
    title=article.find("h1").text
    script=soup.find("div",class_="full-script")
    scriptContentArr=script.find_all("p")
    scriptContent=""
    for j in scriptContentArr:
        scriptContent += j.text+"\n"
    try:
        with open(f"txtFiles\{title}.txt","w", encoding="utf-8") as f:
            f.write(scriptContent)
    except:
        print(f"Couldn't save {title}")
```

```python
url="https://subslikescript.com/movies"
value=1
while True:
    response=requests.get(url + "?page=" + str(value))
    value+=1
    if response.status_code != 200:
        continue
    else:
        soup=BeautifulSoup(response.text,"html.parser")
        parent=soup.find("article",class_="main-article")
        anchor_tags=parent.find_all("a")
        links=[]
        for i in anchor_tags:
            links.append("https://subslikescript.com/"+i['href'])
        for link in links:
            response=requests.get(link)
            soup=BeautifulSoup(response.text,"html.parser")
            article=soup.find("article",class_="main-article")
            title=article.find("h1").text
            script=soup.find("div",class_="full-script")
            scriptContentArr=script.find_all("p")
            scriptContent=""
            for j in scriptContentArr:
                scriptContent += j.text+"\n"
            try:
                with open(f"txtFiles\{title}.txt","w", encoding="utf-8") as f:
                    f.write(scriptContent)
            except:
                print(f"Couldn't save {title}")
    if value==4:
        # for now we are visting till page 3 only.
        break
```
